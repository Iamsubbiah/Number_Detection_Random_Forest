```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
digits=load_digits()                      
```

```python
dir(digits)
```

```python
import matplotlib.pyplot as plt
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
digits.data[1:3]
```

```python
df=pd.DataFrame(digits.data)
df
# meaning all 64 digits look line 0 so its mentioned
```

```python
df["target"]=digits.target
df.head()
```

## train test data splitting

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(['target'],axis=1),digits.target,test_size=0.3,random_state=6)
#                                        x -all table without target     ,y-dependent var   ,testsize     ,randomstate
```

```python
print("trainind data size   =",len(x_train))
print("test data size(30% of training data)  =",len(x_test))

```

# random forest training

```python
from sklearn.ensemble import RandomForestClassifier
# ensembel means using multiple algo to predict output
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

## claculating efficiency of rf model is 96%

```python
rf.score(x_test,y_test)
```

# if you change n_estimators meaning how many trees/branches put togeter gives your model then efficiency changes now its 97.5%

```python
rf=RandomForestClassifier(n_estimators=400)
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```

```python
y_predicted=rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(y_test, y_predicted) 
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7)) 
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted (number of times it found whats input)')
plt.ylabel ('Truth (actual number it was gives as input)')
```
